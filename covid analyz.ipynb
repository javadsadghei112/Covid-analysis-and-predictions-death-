```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('covid19_italy_region.csv')
df
```

```python
df['Country'].value_counts()
```

```python
print(df['RegionCode'].value_counts())
```

```python
plt.figure(figsize=[30,12])
plt.bar(df.RegionName,df.Deaths,);plt.title('death each city');
```

```python
plt.figure(figsize=[30,12])
plt.bar(df.RegionName,df.Recovered,color='red');plt.title('death each city');
```

```python
df.drop(['SNo','Date','RegionName'],axis='columns',inplace=True)
df
```

```python
df['TestsPerformed'].fillna(1,inplace=True)
df
```

```python
df.drop(['Country'],axis='columns',inplace=True)
df
```

```python
X=df.drop(['Deaths'],axis='columns')
y=df.Deaths
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='poly',C=1000,degree=2)
```

```python
svr.fit(X_train,y_train)
```

```python
svr.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
la=LinearRegression()
```

```python
la.fit(X_train,y_train)
```

```python
la.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
ra=RandomForestRegressor(n_estimators=200,max_leaf_nodes=200)
```

```python
ra.fit(X_train,y_train)
```

```python
ra.score(X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gradian=GradientBoostingRegressor(max_depth=3,n_estimators=4,learning_rate=0.8)
```

```python
gradian.fit(X_train,y_train)
```

```python
gradian.score(X_test,y_test)
```
